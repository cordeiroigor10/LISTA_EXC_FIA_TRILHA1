### 1. Ler e exibir um arquivo CSV usando pandas

```
df = spark.read.csv("/FileStore/tables/retail.csv", header=True, sep = ";")
df = df.toPandas()
```

```
df([[u'InvoiceNo', u'StockCode',  u'Description',u'Quantity', 'InvoiceDate', "UnitPrice", "CustomerID", 'Country']], dtype='object')
```

### 2. Selecionar colunas específicas de um DataFrame usando pandas

```
df_1 = df[["Country", "Description", "Quantity"]]
display(df_1)
```

### 3. Filtrar linhas com base em uma condição usando pandas

```
df_2 = df.filter(df['Country']== 'France')
display(df_2)
```

### 4. Agrupar dados com base em uma coluna usando pandas

```
df_3 = df[['Country', "Description", "InvoiceNo"]].groupby(['Country', "Description", "InvoiceNo"])#.reset_index(level =['Country', "Description", "InvoiceNo"])
```
